```python
import pandas as pd
import numpy as np
```

```python
MAX_PROCESSES = 100

df = pd.read_pickle("../game_data_cleaner/game_dfs_dirty/games.pkl")
ratings_totals = pd.DataFrame(df["BGGId"])
ratings_totals["RatingsPages"] = np.ceil(df["NumUserRatings"].astype(int) / 100)
ratings_totals = ratings_totals.sort_values(
    "RatingsPages", ascending=False
).reset_index(drop=True)
# max_ratings_pages = ratings_totals["RatingsPages"].max()
# max_ratings_pages = 2000

max_ratings_pages = int(np.ceil(ratings_totals['RatingsPages'].sum()/MAX_PROCESSES))
print(max_ratings_pages)

ratings_totals.head()
```

```python
group_counter = 1
position = 0
groups_of_games = {}

while len(ratings_totals) > 0:

    chunk_size = 0
    games_and_ratings = {}
    group_entries = []
    

    while max_ratings_pages > chunk_size:
        try:
            bgg_id = ratings_totals.iloc[position]["BGGId"]
            ratings_pages = int(ratings_totals.iloc[position]["RatingsPages"])
            chunk_size += ratings_pages
            games_and_ratings[bgg_id]=ratings_pages
            ratings_totals = ratings_totals.drop(position)
            position += 1

        except:
            break
    
    group_entries.append(games_and_ratings)
    if chunk_size == 0:
        break
    groups_of_games[f"group{group_counter}"] = games_and_ratings
    print(f"group{group_counter} Complete with chunk size {chunk_size}")
    group_counter += 1
```
