```python
new_gamer_synonyms = ['new gamer', 'non gamer', 'newcom', 'newbie', 'novice', 'rookie', 'starter', 'beginner']

# find all entries that contain a word from new_player_synonyms
new_player_entries = [x for x in unique_entries if any([y in x for y in new_gamer_synonyms])]
new_player_entries
```
